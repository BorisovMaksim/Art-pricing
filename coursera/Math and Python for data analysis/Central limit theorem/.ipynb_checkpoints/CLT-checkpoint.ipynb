```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto
from scipy.stats import norm
```

# Распределение Парето

Плотность распределения Парето имеет вид:
$$f(x, b)=\frac{b}{x^{b+1}}$$

Зададим начальные значения переменных:
- size - размер выборки;
- b - параметр распределения Парето:

```python
size = 1000
b = 3.0
```

Сгенерируем выборку объема 1000 из распределения Парето:

```python
r = pareto.rvs(b, size=size)
```

Покажем гистограмму выборки и теоретическую плотность распределения Парето:

```python
x = np.linspace(1, 10, 100)
fig, ax = plt.subplots(1, 1)
ax.plot(x, pareto.pdf(x, b), label='Теоретическая плотность распределения Парето')
ax.legend(loc='best', frameon=False)
ax.hist(r, bins=50, density=True)
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
plt.xlim([0, 10])
plt.show()
```

## Оценка распределения выборочного среднего

Оценим распределение выборочного среднего вашей случайной величины при разных объёмах выборок.

Заведем массив ns, хранящий объемы выборок n. Для каждого n сгенерируем 1000 выборок объема n. Результат запишем в трехмерный массив result.

```python
ns = [5, 10, 50, 100]
result = []
for n in ns:
    current = []
    for i in range(size):
        current.append(pareto.rvs(b, size=n))
    result.append(current)
```

Посчитаем выборочное среднее для всех выборок по формуле:
$$\bar{X}=\frac{1}{n}\sum_{i=1}^n x_i$$
Результат запишем в двумерный массив means, хранящий 1000 выборочных средних для каждого n.

```python
means = []
for i in range(len(ns)):
    current_means = []
    for j in range(size):
        current_means.append(np.mean(result[i][j]))
    means.append(current_means)
```

Посчитаем теоретические математическое ожидание и дисперсию распределения Парето по формулам:
$$E[X]=\frac{b}{b-1}$$
$$D[X]=\left(\frac{1}{b-1}\right)^2 \frac{b}{b-2}$$

```python
mean = b / (b - 1.0)
var = (1.0 / (b - 1.0)) ** 2.0 * (b / (b - 2.0))
print("Мат. ожидание = " + str(mean))
print("Дисперсия = " + str(var))
```

По центральной предельной теореме распределение выборочных средних достаточно хорошо описывается нормальным распределением со следующими параметрами:
$$\bar{X_n}\approx N\left(EX,\frac{DX}{n}\right)$$

Покажем гистограммы распределения выборочных средних для каждого n и плотности нормального распределения, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних.

```python
x = np.linspace(0, 3, 100)
fig, ax = plt.subplots(1, len(ns), sharex=True, sharey=True, figsize=(12,6))
for i in range(len(ns)):
    ax[i].hist(means[i], density=True)
    norm_vs = norm(mean, np.sqrt(var / ns[i]))
    ax[i].plot(x, norm_vs.pdf(x), 'r-')
    ax[i].set_title("n = " + str(ns[i]))
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("f(x)")
fig.tight_layout()
plt.xlim([0, 3])
plt.show()
```

## Выводы

Как можно видеть на графиках, с ростом n форма гистограммы все больше напоминает плотность нормального распределения. Можно утверждать, что с ростом n точность апрокимации распределения возрастает.
