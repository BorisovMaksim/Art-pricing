```python
#f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
import numpy as np
import matplotlib.pyplot  as plt 
```

```python
x_0 = 1
x_1 = 15
a = np.array([[1,1], [15,1]])
b = np.array([np.sin(1/5)*np.exp(1/10) + 5*np.exp(-1/2), np.sin(15/5)*np.exp(15/10) + 5*np.exp(-15/2)])
x = np.linalg.solve(a, b)
x
```

```python
plt.plot([1,15], x)
```
