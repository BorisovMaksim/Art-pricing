```python
import numpy as np
import scipy
import re
```

```python
with open("../coursera/sentences.txt") as read_file:
    init = read_file.readline()
    words = np.array([x for x in re.split('[^a-z]', init.lower()[:-2]) if x!=''])
    data = [words]
    for line in read_file:
        words = [x for x in re.split('[^a-z]', line.lower()[:-2]) if x!='']
        data.append(np.array(words))
```

```python
words_combined = set()
for array in data:
    for string in array: 
        words_combined.add(string)
words_combined
```

```python
dict_all_words = {x[1]:x[0] for x in enumerate(words_combined)}
```

```python
dict_all_words
```

```python
m = [np.array([dict_all_words[word] if word in x else 0 for word in words_combined]) for x in data]
```

```python
np.shape(m)
```

```python
from scipy.spatial import distance
```

```python
a = np.sort([(distance.cosine(m[0], m[i]), i) for i in range(len(m))])
```

```python
ind=np.argsort(a[:,-2])
a[ind]
```

```python
[data[int(col[1])] for col in a[ind]]
```
